```python
import requests
import fake_useragent
import time
```

```python
user_agent = fake_useragent.UserAgent()

headers = {
    'User-Agent': user_agent.random,
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
    'Accept-Language': 'ru-RU,ru;q=0.9',
    'Accept-Encoding': 'gzip, deflate, br',
    'Connection': 'keep-alive',
}

print(f'User-Agent для сессии: {headers}')
```

Создадим сессию через которую будет работать агент

```python
sess = requests.Session()

sess.headers.update(headers)
```

Сначала зайдем на главную страницу, а потом с нее (в рамках 1 сессии) на страницу с вакансиями

```python
try:
    response_main_page = sess.get('https://hh.ru/', timeout=3)

    if response_main_page.status_code == 200:
        print(f'Успешно перещли на главную страницу, ожидаем 1 сек')

        time.sleep(1)  


        get_f1 = sess.get('https://hh.ru/search/vacancy/', headers=headers, timeout=3)
        
        status = get_f1.status_code
        
        if int(status) == 200:
            print(get_f1.text)
        else:
            print(status)
    else:
        print(f'Переход на главную страницу не удался...')
except Exception as ex:
    print(f'Возникла ошибка: {ex}')  
```

```python
link = 'https://hh.ru/search/vacancy/'
browser = 'chrome'


try:
    response = requests.get(url=link, impersonate=browser, default_headers=False)

    status = response.status_code

    if status == 200:
        print(response.text)
    else:
        print(status)
except Exception as ex:
    print(f'Возникла ошибка: {ex}')
```
