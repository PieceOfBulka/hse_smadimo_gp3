```python
import requests
import fake_useragent
```

```python
user_agent = fake_useragent.UserAgent()

headers = {
    'User-Agent': user_agent.random
}

print(f'User-Agent для сессии: {headers}')
```

```python
try:
    get_f1 = requests.get('https://hh.ru/search/vacancy/', headers=headers)
    
    status = get_f1.status_code
    
    if int(status) == 200:
        print(get_f1.text)
    else:
        print(status)
except Exception as ex:
    print(f'Возникла ошибка: {ex}')
```

Для имитации браузера через заголовки и TLS-отпечаток будем использовать специализированную бибилиотеку: **curl_cffi**

https://curl-cffi.readthedocs.io/en/stable/

```python
link = 'https://hh.ru/search/vacancy/'
browser = 'chrome'


try:
    response = requests.get(url=link, impersonate=browser, default_headers=False)

    status = response.status_code

    if status == 200:
        print(response.text)
    else:
        print(status)
except Exception as ex:
    print(f'Возникла ошибка: {ex}')
```
